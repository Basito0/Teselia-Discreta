```python
#Clases
class Pokemon:                      #El nodo
    def __init__(self, nombre):
        self.nombre = nombre        
        self.contactos = []         
        self.posiblesPadres = []    #Todos los que lo tienen como contacto 
        self.padres = []            #Los que lo tienen como contacto y le responde
        self.desconfianza = ""
        self.visited = False    #Variable para scc
        self.assigned = False   #Variable para scc


    def addContacto(self, pokemon):
        if pokemon not in self.contactos:
            self.contactos.append(pokemon)

    def addPadre(self, pokemon):
        if pokemon not in self.posiblesPadres:
            self.posiblesPadres.append(pokemon)

    def setDesconfianza(self, desconfianza):
        self.desconfianza = desconfianza
    
    def calcularConfiables(self):           
        if self.desconfianza == "no":
            self.padres = self.posiblesPadres
            return
        for i in self.posiblesPadres:
            if i in self.contactos:
                self.padres.append(i)


class MapaContacto:                 #Si
    def __init__(self):
        self.pokemones = {}         #Los pokimones
        self.components = {}        #Los scc
    
    def addPokemon(self, pokemon):
        if pokemon not in self.pokemones:
            self.pokemones[pokemon] = Pokemon(pokemon)
    
    def setDesconfianza(self, pokemon, desconfianza):
        self.pokemones[pokemon].setDesconfianza(desconfianza)

    def addContacto(self, padre, hijo):
        if hijo not in self.pokemones:
            self.pokemones[hijo] = Pokemon(hijo)
        self.pokemones[padre].addContacto(self.pokemones[hijo])
        self.pokemones[hijo].addPadre(self.pokemones[padre])

    def conectarConfianza(self):                #Cuando ya todos los pokemones tiene definida su confianza les seteo a los que le responde
        for i in self.pokemones.values():
            i.calcularConfiables()
        
    #SCC
    #Ni idea de lo que pasa aquí, seguí el tutorial que salía en wikipedia xd
    #Referencia: https://en.wikipedia.org/wiki/Kosaraju%27s_algorithm
    def scc(self):
        for i in self.pokemones.values():
            i.visited = False
        self.L = []

        for u in self.pokemones.values():
            self.Visit(u)
        
        for u in self.L:
            self.Assign(u, u)
        return self.components

    def Visit(self, u):
        if u.visited:
            return
        u.visited = True
        for v in u.contactos:
            if not v.visited:
                self.Visit(v)
        self.L.insert(0, u)
        
    def Assign(self, u, root):
        if u.assigned:
            return
        if root.nombre not in self.components:
            self.components[root.nombre] = []
        self.components[root.nombre].append(u)
        u.assigned = True
        for v in u.padres:
            self.Assign(v, root)
    
    #Check if el componente tiene un in vecino que no es del componente:
    def hasExternIn(pokemones):
        for i in pokemones:
            for u in i.padres:
                if u not in pokemones:
                    return True
                
        return False
    
            

#Leer archivo y crear mapa 
mapa = MapaContacto()
fp = open("listing2.dat", 'r')
n = int(fp.readline())
for i in range(n):
    line = fp.readline().strip('\n')
    list = line.split(' ')
    pokimon = list[0]
    mapa.addPokemon(pokimon)
    mapa.setDesconfianza(pokimon, list[-1])
    for j in list[2:-1]:
        mapa.addContacto(pokimon, j)
mapa.conectarConfianza()

#Calcular minimo comun multiplo vola
l = mapa.scc()

#yapu entonces la idea es sacar todos los scc. Entonces sabemos que con que llameN a 1 del componente ya todos van a recibir el mensaje. ahora hay que ver si
#es que hay alGun otro componente que se comunIca coN él. si pasa Eso Entonces no es necesaRio darle el mensaje a nadie de aqui porque se lo podemos dar al del otro
#componente y asi Gg. A los únicos que sí o sí hay que darle el Mensaje son a los que no tIeneN ninGún componente comunicandose con ellos. y eso :v 

#Tonces ya ai calculamos a quien hay que llamar (a Engineer Gaming)
print("Hay que llamar a:")
for i in l:
    if not MapaContacto.hasExternIn(l[i]):
        print(i)



```
